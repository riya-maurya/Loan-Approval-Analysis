```python
df=pd.DataFrame(data)
```

```python
df= df[['Credit_History', 'ApplicantIncome', 'LoanAmount', 'Property_Area', 'Loan_Amount_Term','Loan_Status']]
```

##DATA PREPROCESSING

```python
df
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
df['Credit_History'].fillna(1,inplace=True)
```

```python
df.isnull().sum()
```

```python
# Convert 'LoanAmount' column to numeric, coerce errors to NaN
df['LoanAmount'] = pd.to_numeric(df['LoanAmount'], errors='coerce')
# Now, fill NaN values with the median
df['LoanAmount'].fillna(df['LoanAmount'].median(), inplace=True)
```

```python
df.isnull().sum()
```

```python
df['Loan_Amount_Term'].fillna(df['Loan_Amount_Term'].median(),inplace=True)
```

```python
df.isnull().sum()
```

```python
df['Property_Area'].unique()
```

```python
df['Property_Area'].replace({'Urban':1,'Semiurban':2,'Rural':3},inplace=True)
df['Loan_Status'].replace({'Y':1,'N':0},inplace=True)
```

###Testing

```python
from sklearn.model_selection import train_test_split
```

```python
X_test, X_train, y_test, y_train = train_test_split(df[['ApplicantIncome','Credit_History','LoanAmount','Property_Area','Loan_Amount_Term']], df['Loan_Status'], test_size=0.2, random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dl=DecisionTreeClassifier()
```

```python
dl.fit(X_train,y_train)
```

```python
dl.score(X_test,y_test)
```
